# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
        return abs(x) * (x > 0)
def dReLU(x):
        return ( 1 * (x > 0))
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, ReLU(x), 'r',label='linspace(-10,10,10)')
plt.plot(x, dReLU(x), 'b',label='linspace(-10,10,10)')
plt.grid()
plt.title('ReLU & dReLU Function')
plt.show()
```
